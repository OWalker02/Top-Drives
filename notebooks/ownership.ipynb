```python
def split_tune(tune: int):
    # Convert to string
    tune_str = str(tune)
    # Pad with leading zeros to ensure it's 3 digits, then reduce to last 3 digits
    tune_str = tune_str.zfill(3)[-3:]
    # Get each tune
    eng = int(tune_str[0])
    wei = int(tune_str[1])
    cha = int(tune_str[2])
    return eng, wei, cha
```

```python
def calc_unowned_penalty(rq, car_version, prize, eng_up, reduction_tags):
    # Check reduction tags
    reduce = any(reduction_tags)
    
    penalties = {
        80: 100000,
        65: 20000,
        50: 3000,
        40: 750,
        30: 300,
        20: 100,
        0: 150
    }
    version_multipliers = {
        80: 5,
        65: 3,
        50: 2,
        40: 1,
        30: 1,
        20: 1,
        0: 1
    }
    upgrade_penalties = {
        80: 10000,
        65: 300,
        50: 60,
        40: 20,
        30: 10,
        20: 2,
        0: 4
    }

    # If prize car, return 10000000
    # Else, return pen = rarity_pen + upgrade_pen * version_mult
    #     rarity_pen is a base penalty determined by the rarity of the car
    #     upgrade_pen is based on the rarity, if car has any reduction tags, and if it is a prize car
    #     version_mult is to multiply the upgrade pen for duplicates

    if prize == 1:
        return 1e7
    
    # Get the rq_key
    rq_key = 0
    rq_upper = 115
    for rq_lim in penalties.keys():
        if rq >= rq_lim:
            rq_key = rq_lim
            break
        else:
            rq_upper = rq_lim - 1
    
    # Get pen components
    rq_pen = rq_upper - rq
    base_pen = penalties[rq_key]
    version_mult = version_multipliers[rq_key]
    indiv_up_pen = upgrade_penalties[rq_key]
    up_pen = indiv_up_pen * 5

    # Calc pen
    if eng_up == 1:
        pen = base_pen
    else:
        pen = base_pen + up_pen * (1 + version_mult*car_version)

    # Reductions
    if reduce:
        pen = round(pen / 2)
    
    # Add RQ pen
    pen += rq_pen

    return pen



def calc_penalty(rq, engine_up, weight_up, chassis_up, own_eng, own_wei, own_cha, car_version, prize, reduction_tags):
    # Check reduction tags
    reduce = any(reduction_tags)

    upgrade_penalties = {
        80: 10000,
        65: 300,
        50: 60,
        40: 20,
        30: 10,
        20: 2,
        0: 4
    }
    version_multipliers = {
        80: 5,
        65: 3,
        50: 2,
        40: 1,
        30: 1,
        20: 1,
        0: 1
    }

    # Change all 0 upgrades to 1
    if own_eng == 0:
        own_eng = 1
    if own_wei == 0:
        own_wei = 1
    if own_cha == 0:
        own_cha = 1

    # Calculate differences between each data tune and owned tune
    eng_diff = engine_up - own_eng
    wei_diff = weight_up - own_wei
    cha_diff = chassis_up - own_cha

    # Get the rq_key
    rq_key = 0
    rq_upper = 115
    for rq_lim in upgrade_penalties.keys():
        if rq >= rq_lim:
            rq_key = rq_lim
            break
        else:
            rq_upper = rq_lim - 1
    
    # Get pen components
    rq_pen = rq_upper - rq
    version_mult = version_multipliers[rq_key]
    indiv_up_pen = upgrade_penalties[rq_key]

    # Check which type of pen to calculate
    min_diff = min(eng_diff, wei_diff, cha_diff)
    max_diff = max(eng_diff, wei_diff, cha_diff)

    # If any are less than zero, the car is upgraded past this tune so treat as unowned
    # If all are exactly zero, the car is upgraded
    # Else, the car is upgradeable

    if min_diff < 0:
        pen = calc_unowned_penalty(rq, car_version, prize, engine_up, reduction_tags)
    elif min_diff == 0 and max_diff == 0:
        pen = 0
    else:
        ups_togo = eng_diff + wei_diff + cha_diff
        up_pen = ups_togo * indiv_up_pen
        pen = up_pen * (1 + version_mult*car_version)

    # Add RQ pen
    pen += rq_pen
    
    return pen
```

```python
def calc_penalty(rq, engine_up, weight_up, chassis_up, own_eng, own_wei, own_cha, car_version, prize, reduction_tags):
    # Check reduction tags
    reduce = any(reduction_tags)

    upgrade_penalties = {
        80: 10000,
        65: 300,
        50: 60,
        40: 20,
        30: 10,
        20: 2,
        0: 4
    }
    version_multipliers = {
        80: 5,
        65: 3,
        50: 2,
        40: 1,
        30: 1,
        20: 1,
        0: 1
    }

    # Change all 0 upgrades to 1
    if own_eng == 0:
        own_eng = 1
    if own_wei == 0:
        own_wei = 1
    if own_cha == 0:
        own_cha = 1

    for rq_lim, base_pen in upgrade_penalties.items():
        if rq >= rq_lim:
            version_mult = version_multipliers[rq_lim]
            break

    # Calculate differences between each data tune and owned tune
    eng_diff = engine_up - own_eng
    wei_diff = weight_up - own_wei
    cha_diff = chassis_up - own_cha

    # Initialise pen
    pen = 666

    # Check which type of pen to calculate
    min_diff = min(eng_diff, wei_diff, cha_diff)
    max_diff = max(eng_diff, wei_diff, cha_diff)

    # If any are less than zero, the car is upgraded past this tune so treat as unowned
    # If all are exactly zero, the car is upgraded
    # Else, the car is upgradeable

    if min_diff < 0:
        pen = calc_unowned_penalty(rq, car_version, prize, engine_up, reduction_tags)
    elif min_diff == 0 and max_diff == 0:
        pen = 0
    else:
        prev_rq = 120
        for rq_lim, base_pen in upgrade_penalties.items():
            if rq >= rq_lim:
                # Calculate rq penalty (upper rq limit for rarity - rq)
                rq_pen = prev_rq - 1 - rq
                # Calculate the penalty based on the number of upgrades
                eng_pen = eng_diff * base_pen
                wei_pen = wei_diff * base_pen
                cha_pen = cha_diff * base_pen
                ups_pen = eng_pen + wei_pen + cha_pen
                # Add the version multiplier
                mult_pen = version_mult * car_version
                pen = (mult_pen + 1) * ups_pen
                # Reduce penalty for prize cars
                if prize == 1:
                    pen = round(pen / 2)
                # Reduce penalty for Marilyn's Collection
                if reduce:
                    pen = round(pen / 2)
                # Add rq pen
                pen += rq_pen
                break
            prev_rq = rq_lim
    return pen
```
