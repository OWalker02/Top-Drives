## What each string contains

- components_js:
    - Tracks / surfaces for add tracks list (starts at "bo=JSON.parse")
    - JSON list of campaign matches (starts at "xo=JSON.parse")
- index_js
    - JSON list of all car info dicts (starts at "Q=JSON.parse")
    - List of all mixed surface types' percentages (starts at "o1=")
    - JSON list of all rids (starts at "Ze=JSON.parse")
    - List of rarity hex codes (starts at "B2=") and RGB codes (starts at "G2=")
    - List of js_vars to understandable strings (including tracks e.g. drag100b: 0-100-0 mph) (starts at "Ib=")
