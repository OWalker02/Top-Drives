# Imports

```python
import json

from selenium.webdriver import Firefox, FirefoxOptions

from config.paths import RAW_SCRAPED_JSON_PATH
from config.scraping import CLICK_OFF_COORDS
from src.scraping.scrapers import TDRScraper
from src.scraping.scraping import update_db
```

# Scraping

```python
options = FirefoxOptions()
driver = Firefox(options=options)
```

```python
scraper = TDRScraper(
    driver,
    CLICK_OFF_COORDS,
    {"tags": ["Crown Pursuit"], "rarities": ["D"]},
    scrape_new=True,
    test_mode=False,
)
```

```python
scraper.scrape()
```

```python
db = update_db(scraper)
```

```python
with open(RAW_SCRAPED_JSON_PATH, "w") as f:
    json.dump(db, f)
```
